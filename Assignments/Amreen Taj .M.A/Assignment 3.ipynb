# 5. Check for Missing values and deal with them.

```python
data.isna().any()
```

```python
data['Rings'].fillna(data['Rings'].mean(),inplace=True)
data
```

# 6. Find the outliers and replace them outliers

```python
sns.boxplot(data.Diameter)
```

```python
dat=data[data['Diameter']<0.10]
sns.boxplot(y='Diameter',data=dat)
```

```python
Q1=data.Diameter.quantile(0.25)
Q2=data.Diameter.quantile(0.75)
IQR=Q2-Q1
print(IQR)
```

```python
data=data[~((data.Diameter<Q1-1.5*IQR))|(data.Diameter>(Q2+1.5*IQR))]
data
```

# 7. Check for Categorical columns and perform encoding.

```python
data['Sex'].replace({'M':1,'F':0},inplace=True)
data.head()
```

# 8. Split the data into dependent and independent variables.

```python
dmain=pd.get_dummies(data,columns=['Height'])
dmain
```

```python
y=dmain['Diameter']
y
```

```python
x=dmain.drop(columns='Diameter',axis=1)
x.head()
```

# 9. Scale the independent variables

```python
x=data.iloc[:,6:7].values
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
x=std.fit_transform(x)
x
```

# 10. Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

# 11.Build the Model

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

# 12.Train the Model

```python
plt.scatter(x_train,y_train,color='yellow')
plt.plot(x_train,regressor.predict(x_train))
```

# 13.Test the Model

```python
y_pre=regressor.predict(x_test)
y_pre
```

# 14. Measure the performance using Metrics.

```python
from sklearn.metrics import r2_score
accuracy=r2_score(y_test,y_pre)
accuracy
```
